# Data Preparation

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import sklearn.metrics as metrics
```

```python
#Split Train and Test Set
X = new_data.drop(columns=['Level'])
y = np.log(new_data['Level'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
X.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from sklearn.neighbors import KernelDensity
```

```python
#mencari nilai k dengan cross validation untuk Kneighbors
k_values = list(range(1, 31))  # Coba nilai k dari 1 hingga 30
cv_scores = []

for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='r2')
    cv_scores.append(scores.mean())

# Menampilkan hasil evaluasi akurasi untuk setiap nilai k
for k, score in zip(k_values, cv_scores):
    print(f"K = {k}, R-Squared = {score}")

# Menentukan nilai k terbaik
best_k = k_values[np.argmax(cv_scores)]
print(f"\nNilai k terbaik: {best_k}")
```

###### Noted
Semakin tinggi nilai R-Squared semakin baik model yang dihasilkan.
Semakin tinggi nilai K model yang dihasilkan akan membentuk pola yang terlalu umun dan tidak dapat mengatasi pola yang kompleks. Jika dilihat dari hasil cross validation, semakin tinggi nilai K maka semakin menurun nilan R-Squarednya. Setelah dicoba hasil akurasi dari K1 - K9 yang tetap menghasilkan RMSE pada nilai 0 adalah K1-K6. Oleh karena itu diputuskan untuk menggunakan nilai k = 6 untuk tetap menjaga akurasi dan membuat model yang bisa membaca pola yang lebih kompleks 

```python
# Siapkan dataframe untuk analisis model
report = pd.DataFrame(index=['train_mse', 'test_mse'], 
                      columns=['Kneighbors', 'Decision Tree Regressor', 'Extra Tree Regressor'])
```

```python
k = 6  
K = KNeighborsRegressor(n_neighbors=k)
K.fit(X_train, y_train)


D = DecisionTreeRegressor()
D.fit(X_train, y_train)

E = ExtraTreesRegressor()
E.fit(X_train, y_train)
```

# Evaluasi

```python
final_report = {'Model_Name': [], 'mse': [], 'r2': []}
pred = K.predict(X_test)
mse = mean_squared_error(y_true=y_test, y_pred=pred)
r2 = r2_score(y_test, pred)
final_report['Model_Name'].append('Kneighbors')
final_report['mse'].append(mse)
final_report['r2'].append(r2)
pred = D.predict(X_test)
mse = mean_squared_error(y_true=y_test, y_pred=pred)
r2 = r2_score(y_test, pred)
final_report['Model_Name'].append('Decision Tree Regressor')
final_report['mse'].append(mse)
final_report['r2'].append(r2)
pred = E.predict(X_test)
mse = mean_squared_error(y_true=y_test, y_pred=pred)
r2 = r2_score(y_test, pred)
final_report['Model_Name'].append('Extra Tree Regressor')
final_report['mse'].append(mse)
final_report['r2'].append(r2)
```
